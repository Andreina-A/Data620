```python
#Import packages for Krackhardt Kite
import networkx as net
import matplotlib.pyplot as plt

#Started by creating a graph
g= net.Graph()

#Added egdes
#Edges created with names provided.
Edge= [("Andre","Carol"), ("Andre","Fernando"), ("Andre", "Diane"), ("Andre","Beverly"),
       ("Carol","Fernando"), ("Carol", "Diane"),("Heather", "Garth"),
        ("Heather","Fernando"), ("Fernando", "Garth"), ("Heather","Ike"), ("Ike","Jane"),
       ("Diane","Garth"), ("Diane","Fernando"), ("Ed", "Diane"), ("Diane","Beverly"),
      ("Garth","Ed"), ("Beverly", "Ed"), ("Beverly","Garth")]
g.add_edges_from(Edge)


#Visualize graph
net.draw(g, with_labels=True)
plt.title("Krackhardt Kite's Graph")
plt.show()

```
